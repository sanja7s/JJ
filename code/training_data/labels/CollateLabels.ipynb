```python
import pandas as pd
```

```python
all_data = pd.DataFrame()
cnt = 0
for city in ["bologna", "milano", "roma", "firenze", "palermo"]: # ["bologna", "milano", "roma", "firenze", "palermo"]
    data = pd.read_csv("../../../preprocessed/OLD/labels/"+city+"_imagelet_labels.csv")
    cnt += len(data)
    data["city"] = city
    all_data = pd.concat([all_data, data], sort=False)
```
