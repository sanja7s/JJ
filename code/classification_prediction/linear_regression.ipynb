```python
import pandas as pd
import statsmodels.api as sm
from sklearn.preprocessing import StandardScaler
```

```python
CITY = "milano"
data_dir = "../../preprocessed/" 
features_dir = data_dir + "features/" + CITY + "/"
labels_dir = data_dir + "labels/" + CITY + "/"

PCA_components = 32

standardize_features = True

feature_size = 2048

network_type='VGG'

features_columns = ["PCA"+str(i) for i in range(PCA_components)]
```

```python
label_columns = ["hType_mix", "num_intersect", "bld_avg_age", "emp_rat_num",\
				"LUM5_single",	"RNR_nres", "mdist_smallparks", "nig_rat_daily",\
				"nig_rat_daily3", "mdist_nres_daily", "num_community_places", \
				"num_community_places_poi", "avg_block_area", "sphi", \
				"enterprises_empl_size", "pop_rat_num",  \
				"emp_rat_pop", "bld_rat_area", "den_nres_daily",\
				"mdist_parks", "den_nres_non-daily", "mdist_railways",\
				"mdist_highways", "mdist_water"]
```

```python
def read_normalized_labels_features(network_type='VGG'):
    if network_type == 'ResnNet':
        return pd.read_csv(features_dir+"Resnet50/df_ResNet50_feat2048_pca32_labels_features.csv")
    else:
        return pd.read_csv(features_dir+"VGG16/df_VGG16_feat2048_pca32_labels_features.csv")
```

```python
data = read_normalized_labels_features()
```

```python
data.head()
```

```python
def predict_label_i(label="label_hType_mix"):
    
    target = data[["image", label]]
    features = data[features_columns]

    X = features.values
    y = target[label].values
    
    X = sm.add_constant(X)  
    model = sm.OLS(y,X)
    
    results = model.fit()
    
    print ("LABEL ", label)
    print_model = results.summary()
    print (print_model)
    
```

```python
predict_label_i()
```

```python
for col in label_columns:
    label = "label_" + col
    print (label)
    predict_label_i(label)
```
